```python
df.drop_duplicates(inplace=True) #inplace:> make the change directly in df itself, don't create a new dataframe.
df.shape
```

```python
#here age in object form so what iam doing is changing it to numeric form
df['Age'].describe()
```

```python
#first convert it to numeric form 
df['Age'] = pd.to_numeric(df['Age'], errors='coerce') #coerce:>if any value can't be converted to number, make it NaN instead of crashing.
```

```python
df['Age']=df['Age'].fillna(df['Age'].median())  #fill NaN with median
```

```python
df['Age'] = df['Age'].astype(int)  #then change dtype to integer after removing missing values
```

```python
df=df[(df['Age'] >= 1) & (df['Age'] <= 100)] #check for min max and Keep only the rows where Age is between 1 and 100
```
